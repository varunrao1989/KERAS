```python
import pandas as pd
import numpy as np

from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

# Solution for SECTION A

#### Import the data and make sure it is clean

```python
df = pd.read_csv(r"C:\Users\The Godfather\Desktop\concrete_data.csv")
df.head()
```

```python
df.shape
```

```python
df.describe()
```

#### Split the Factors(X) and Response(y)

```python
X = df[["Cement","Blast Furnace Slag", "Fly Ash", "Water", "Superplasticizer", "Coarse Aggregate", "Fine Aggregate", "Age"]] # all columns except Strength
y = df["Strength"] # Strength column
```

#### Without normalization 

#### import Keras, build and compile the model

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
n_cols = X.shape[1] # number of predictors for section A and B
```

```python
model = Sequential()
model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
model.add(Dense(1))

#complile
model.compile(optimizer='adam', loss='mean_squared_error')
```

#### Train the model, make predictions and evaluate MSE x50 times

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = []

for i in range (0,50):   
    X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3)
    model.fit(X_train, y_train, epochs=50, verbose=2)
    yhat_test = model.predict(X_test)
    m = mean_squared_error(y_test, yhat_test)
    mse.append(m)

print(mse)
```

```python
mean = (sum(mse)/len(mse))
print("Mean of the MSEs without normalization is: %.2f" % mean)

import statistics
stdev = statistics.pstdev(mse)
print("Standard deviation of the MSEs without normalization is:  %.2f" % stdev)
```

# SOLUTION for SECTION B

### Normalize X

```python
X_norm = (X - X.mean())/X.std()
X_norm.head()
```

```python
mse = []

for i in range (0,50):   
    X_train, X_test, y_train, y_test = train_test_split( X_norm, y, test_size=0.3)
    model.fit(X_train, y_train, epochs=50, verbose=2)
    yhat_test = model.predict(X_test)
    m = mean_squared_error(y_test, yhat_test)
    mse.append(m)

print(mse)
```

```python
mean = (sum(mse)/len(mse))
print("Mean of the MSEs with normalization is: %.2f" % mean)

import statistics
stdev = statistics.pstdev(mse)
print("Standard deviation of the MSEs with normalization is:  %.2f" % stdev)
```

#### The mean seems to have decreased (which is good) , but the standard deviation seems to have increased from A.

# SOLUTION for SECTION C

```python
mse = []

for i in range (0,50):   
    X_train, X_test, y_train, y_test = train_test_split( X_norm, y, test_size=0.3)
    model.fit(X_train, y_train, epochs=100, verbose=2)
    yhat_test = model.predict(X_test)
    m = mean_squared_error(y_test, yhat_test)
    mse.append(m)

print(mse)
```

```python
mean = (sum(mse)/len(mse))
print("Mean of the MSEs with normalization and 100 epochs is: %.2f" % mean)

import statistics
stdev = statistics.pstdev(mse)
print("Standard deviation of the MSEs with normalization and 100 epochs is:  %.2f" % stdev)
```

#### The mean and the standard deviation seems to have decreased from B. Step toward the right direction. More epochs led to better back propagation and better optimization of weights.
